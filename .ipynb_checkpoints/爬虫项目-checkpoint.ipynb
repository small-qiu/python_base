#### 链家

```python
import requests
import re
import time
from lxml import etree
from bs4 import BeautifulSoup
from pyquery import PyQuery as pq
import numpy as np
import pandas as pd

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
pre_url = 'https://bj.lianjia.com/ershoufang/pg'
```

```python
# 捕获网页信息,网页源代码
def get_html(url,headers):
    
    respones = requests.get(url,headers =headers)
    respones.encoding = 'utf-8'
    time.sleep(1)
    
    return respones.text

# 解析网页，提取网页信息
def get_text(text):
    # xpath
    html = etree.HTML(text)
    house_list = html.xpath('//li//div[@class="address"]/div/text()|//li//div[@class="priceInfo"]   \
                            /div[@class="totalPrice"]/span/text() |//li//div[@class="positionInfo"]/a/text() ') 
    
    
    
    ### 三山新新家园 - 后沙峪,1室1厅 | 58.32平米 | 东 西 | 简装 | 中楼层(共5层) | 板楼,228
    
    return house_list

# 信息处理
def text_to_pd(house_list):
    gs = np.array(house_list) # 列表变nparray
    gs = gs.reshape(-1,4) # nparray变矩阵
    gs = pd.DataFrame(gs) # 矩阵变表格
    # 一二行地址相连
    gs['address']=gs[0]+gs[1]
    for i in range(gs.shape[1]):
        r = gs[2][i].split('|')
        if len(r)<7:
            continue
        gs['房型'] = r[0]
        gs['大小'] = r[1]
        gs['方位'] = r[2]
        gs['装修'] = r[3]
        gs['楼层'] = r[4]
        gs['建设'] = r[5]
        gs['楼型'] = r[6]
    gs['价格/w'] = gs[3]
    gs = gs.drop([0,1,2,3],axis=1)
    pd_to_file(gs)
    
# 写入文件
def pd_to_file(gs):
    pre_pd = pd.read_excel('house_list.xls',encoding='utf-8')
    gs = pd.concat([pre_pd,gs],axis=0,ignore_index=True)
    gs.to_excel('house_list.xls',encoding='utf-8')
```

```python
pd.DataFrame().to_excel('house_list.xls')
for pg in range(1,101):
    res = get_html(pre_url+str(pg),headers = headers)
    g =  get_text(res)
    g = text_to_pd(g)
print('end')
```

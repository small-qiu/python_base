```python
import tkinter as tk
import numpy as np
window = tk.Tk()
# ... all window parts
window.title('取数游戏')
window.geometry('750x400')
who = [' A']
e = tk.Entry(window, width=80)
e.grid(row=0, column=0, columnspan=2)
n = list(np.random.randint(0,100,4))
A = []
B = []
def my_func():
    if who[-1] == ' A':
        e.insert("end", who[-1]+': ')
        who.append(' B')
    else:
        e.insert("end", who[-1]+': ')
        who.append(' A')
    
def move():
    delete = int(e.get().split()[-1])
    n.remove(delete)
    if len(A) > len(B): 
        B.append(delete)
    else: A.append(delete)
    var = 'A='+str(A)+'B='+str(B)
    lb.insert("end", var)
    lb.insert("end","剩余"+str(n))
def printSomething():
    # if you want the button to disappear:
    # button.destroy() or button.pack_forget()
    lb.insert("end", "每次取最左或者最右端一个数，最后累加和大的获胜")
    lb.insert("end", "本次随机生成数值为：")
    lb.insert("end", str(n))
    #this creates a new label to the GUI 
b = tk.Button(window, text="游戏规则", command=printSomething) 
b.grid(row=2, column=1)
b = tk.Button(window, text='选数', width=20, height=3,
command=my_func)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   
b.grid(row=1, column=0)
lb = tk.Listbox(window, width=40)
lb.grid(row=2, column=0)
b = tk.Button(window, text='确定', width=20, height=3,command=move)
b.grid(row=1, column=1)


def final():
    SumA = sum(A)
    SumB = sum(B)
    lb.insert("end",'A:'+str(SumA)+'B:'+str(SumB))
    if SumA > SumB :
        lb.insert("end",'A win')
    else:
        lb.insert("end",'B win')
b = tk.Button(window, text='结束', width=20, height=3,command=final)
b.grid(row=3, column=0)
# canvas = tk.Canvas(window, height=200, width=500)
# image_file = tk.PhotoImage(file="data/welcome.gif")
# image = canvas.create_image(0, 0, anchor='nw', image=image_file)
# canvas.grid(row=3, column=0, columnspan=2)
window.mainloop()
```
